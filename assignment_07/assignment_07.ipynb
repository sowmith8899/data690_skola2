**Defination of Indicators :**
- **NY.GDP.PCAP.PP.CD** -- GDP per capita, PPP (current international $)
- **SP.DYN.LE00.IN** -- Life expectancy at birth, total (years)
- **SH.STA.SUIC.P5** -- Suicide mortality rate (per 100,000 population)
- **SP.POP.TOTL** -- Population, total

```python
df1 = df.loc[df["Country Code"].isin(["NLD","IND","NIC","PER","NRU","IDN","TTO","CRI"])]
df1
```

Selecting GDP as indicator for 2021

```python
gdp1 = df1.query('indicator == "NY.GDP.PCAP.PP.CD" and Year == 2021')
gdp1
```

```python
fig = px.bar(gdp1, x='Country Name', y='value',color='Country Name',title='GDP rate of selected 8 countires for 2021 using bar')
fig.show()
```

```python
fig = px.pie(gdp1, names='Country Name', values='value',color='Country Name',title='GDP rate of selected 8 countires for 2021 using pie')
fig.show()
```

```python
fig = px.box(gdp1, x='Country Name', y='value',color='Country Name',title='GDP rate of selected 8 countires for 2021 using pie')
fig.show()
```

```python
gdp2 = df1.query('indicator == "NY.GDP.PCAP.PP.CD"')
gdp2
```

```python
fig = px.line(gdp2, x='Year', y='value',color='Country Name',title='GDP rate of selected 8 countires for all years using line')
fig.show()
```

```python
df5 = pd.pivot_table(df, values = 'value', index=['Country Code', 'Country Name','Region','Year'], columns = 'indicator').reset_index()
df5.dropna(inplace=True)
df5
```

```python
df5.groupby(['Region']).mean().plot(figsize = (20,9),title = 'Comparision b/w Regions and Indicators')
```
