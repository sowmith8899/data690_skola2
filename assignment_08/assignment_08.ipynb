### Cleanse the data to make the year, population, cost, cost per person numerical type

#### Code functions to cleanse the columns. One function for one column

**Years** 

```python
def cleanse_year(a_string):
    return int(a_string.replace("*",""))
```

```python
df["Year Int"] = df["Census Year"].apply(cleanse_year)
df.info()
df.head(2)
```

**Population**

```python
def cleanse_pop(a_string):
    return int(a_string.replace(",",""))
```

```python
df["Population Int"] = df["Total Population"].apply(cleanse_pop)
df.info()
df.head(2)
```

**Census Cost**

```python
def cleanse_census(a_string):
  if "Billion" in a_string:
    a_string=a_string.replace(",","").replace("$","").replace("Billion","")
    a_string=(float(a_string))*1000000000
    return a_string
  else:
    return float(a_string.replace(",","").replace("$",""))
```

```python
df["Census Int"] = df["Census Cost"].apply(cleanse_census)
df.info()
df.head(2)
```

**Average Cost Per Person**

```python
def cleanse_cost(a_string):
  if "cents" in a_string:
    a_string=a_string.replace("cents","")
    a_string=(float(a_string))/100
    return a_string
  else:
    return float(a_string.replace("$",""))
```

```python
df["Avg Cost Int"] = df["Average Cost Per Person"].apply(cleanse_cost)
df.info()
df.head(2)
```
